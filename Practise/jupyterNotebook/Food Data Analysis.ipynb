```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn
import scipy as sp
```

```python
food_production = pd.read_csv('Africa Food Production (2004 - 2013).csv')
food_production.head()
```

```python
country = food_production.groupby(['Country', 'Year']).sum()
country.head()
```

```python
line_plot = sns.lineplot(x=food_production["Year"], y=food_production["Value"], hue=food_production["Country"], palette = 'bright', ci = None, data= food_production)
line_plot.set_yscale('log')
line_plot.legend(loc='upper right', bbox_to_anchor=(1.4, 1))
fig = plt.gcf()
fig.set_size_inches(12, 8)
```

```python
food_supply = pd.read_csv('Africa Food Supply (2004 - 2013).csv')
food_supply.head()
```

```python
line_plot = sns.lineplot(x=food_supply["Year"], y=food_supply["Value"], hue=food_supply["Country"], palette = 'bright', ci = None, data= food_supply)
line_plot.set_yscale('log')
line_plot.legend(loc='upper right', bbox_to_anchor=(1.4, 1))
fig = plt.gcf()
fig.set_size_inches(12, 8)
```
