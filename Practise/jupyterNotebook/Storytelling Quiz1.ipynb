```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Food_production = pd.read_csv('Africa Food Production (2004 - 2013).csv')
Food_production.head()
```

```python
Food_production.describe()
```

```python
Food_Country = Food_production.groupby(['Country', 'Year']).sum()
Food_Country.head(10)
```

```python
Country_group_p = Food_production.groupby(['Country'])
Country_group_p['Value'].describe().sort_values(by='max', ascending=False)
```

```python
Production_yearly_value = pd.pivot_table(Food_production, index = 'Country', columns = 'Year', values = 'Value')
Production_yearly_value
```

```python
Food_supply = pd.read_csv('Africa Food Supply (2004 - 2013).csv')
Food_supply.head()
```

```python
Food_supply.describe()
```

```python
consumption_Country = Food_supply.groupby(['Country', 'Year']).sum()
consumption_Country.head(10)
```

```python
Country_group_s = Food_supply.groupby(['Country'])
Country_group_s['Value'].describe().sort_values(by='max', ascending=False)
```

```python
Supply_yearly_value = pd.pivot_table(Food_supply, index = 'Country', columns = 'Year', values = 'Value')
Supply_yearly_value
```
